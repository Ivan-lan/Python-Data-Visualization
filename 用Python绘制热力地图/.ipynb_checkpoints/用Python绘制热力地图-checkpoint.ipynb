## 练习1：画世界地图

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%matplotlib inline  

plt.figure(figsize=(16,8))  # 定义了图的大小
m = Basemap()  # 创建一个地图
m.drawcoastlines() # 把海岸线画上
plt.show() # 显示这个地图
```

## 练习3:画中国地图

在创建Basemap时指定一下经纬度

Basemap参数：

llcrnrlon	所需地图域（度）的左下角经度。

llcrnrlat	所需地图域左下角的纬度（度）。

urcrnrlon	所需地图域（度）的右上角经度。

urcrnrlat	所需地图域右上角的纬度（度）。

lat_0	中心纬度（y轴原点） - 由所有投影使用。

lon_0	中央子午线（x轴原点） - 由所有投影使用。

```python
plt.figure(figsize=(16,8))  # 定义了图的大小
m = Basemap(llcrnrlon=73, llcrnrlat=18, urcrnrlon=135, urcrnrlat=53)      # 创建一个地图
m.drawcoastlines() # 把海岸线画上
m.drawcountries(linewidth=1.5)  # 画上国家
plt.show() # 显示这个地图
```

上图看上去有点变形，因为没有添加任何投影，Basemap提供24种不同的投影方式，比较常用的是兰勃特投影  projection='lcc'

## 练习8：给中国地图加上颜色

```python
from matplotlib.patches import Polygon

plt.figure(figsize=(16,8))  # 定义了图的大小
m = Basemap(llcrnrlon=77, llcrnrlat=14, urcrnrlon=140, urcrnrlat=51, projection='lcc', lat_1=33, lat_2=45, lon_0=100)      # 创建一个地图
m.drawcoastlines() # 把海岸线画上
m.drawcountries(linewidth=1.5)  # 画上国家
m.readshapefile('gadm36_CHN_shp\\gadm36_CHN_1', 'states', drawbounds=True)  # 文件为shp，但不用加后缀.shp
# m.readshapefile('gadm36_CHN_shp\\gadm36_CHN_2', 'states', drawbounds=True)  # 2 分得更细致


ax = plt.gca()  # gca就是Get Current Axes的缩写，实际上就是要获得当前图形的座标轴
# 开始一个循环，把图形文件中各个省的多边形取出来，给它一个颜色，统一放上红色r，然后把这个多边形放在我们图形的座标轴上
for nshape, seg in enumerate(m.states):  
    poly = Polygon(seg, facecolor='r')
    ax.add_patch(poly)
    
# 需要再把台湾的颜色加上去
m.readshapefile('gadm36_TWN_shp\\gadm36_TWN_0', 'taiwan', drawbounds=True)  # 加上台湾省的    
for nshape, seg in enumerate(m.taiwan):
    poly = Polygon(seg, facecolor='r')
    ax.add_patch(poly)   
    
plt.show() # 显示这个地图
```

## 练习9：用中国分省人口数据画热力地图

各省人口数据从国家统计局下载 http://www.stats.gov.cn/tjsj/pcsj/rkpc/6rp/indexce.htm  整理出地区和人口数据。

```python
import pandas as pd
df = pd.read_excel('chnpop.xls')  # 读取各省人口数据
df.head()
```

完整代码如下：

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.colors import rgb2hex
import pandas as pd
import numpy as np

plt.figure(figsize=(16,8))  # 定义图的大小
m = Basemap(
    llcrnrlon=77,
    llcrnrlat=14,
    urcrnrlon=140,
    urcrnrlat=51,
    projection='lcc',
    lat_1=33,
    lat_2=45,
    lon_0=100
)  # 创建地图 

m.drawcountries(linewidth=1.5) # 画国家

m.drawcoastlines(color=coastline_color)  # 画海岸线

m.readshapefile('gadm36_CHN_shp\\gadm36_CHN_1', 'states', drawbounds=True)  # 读取中国大陆的省界区划  （读入形状文件，可选择在地图上绘制边界）

df = pd.read_excel('chnpop.xls') # 读取分省人口数据
df['省名'] = df.地区.str[:2]   # 只取省名前两个字
df.set_index('省名', inplace=True)  # 将省名设置为索引

statenames=[]
colors={}
cmap = plt.cm.YlOrRd  # 需要选择一个调色板，即色彩映射表colormap，https://matplotlib.org/examples/color/colormaps_reference.html
vmax = 100000000  # 一个亿
vmin = 3000000  # 300万

for shapedict in m.states_info:  # 形状列表，理解见上
    statename = shapedict['NL_NAME_1']  #  例子：'NL_NAME_1': '北京|北京',
    p = statename.split('|') # 分割字符串得到列表
    # print(p)  从打印结果来看，分割后的列表里的元素：有的一个，有的两个，有区分繁体字和简体字，所以才做以下的判断
    if len(p) > 1:   
        s = p[1]
    else:
        s = p[0]  
    s = s[:2]  # 只取省名的前两个字
    if s == '黑龍':  # 将黑龙繁体转为简体
        s = '黑龙'
    statenames.append(s)  # 添加进省名列表 statenames=[]
    
    pop = df['人口'][s] # 取出对应省份的人口数
    colors[s] = cmap(np.sqrt((pop - vmin) / (vmax - vmin)))[:3]  # 把每个省的人口数据映射到colormap，sqrt() 方法返回数字x的平方根。
# colors={} 字典
    
ax = plt.gca()  # gca就是Get Current Axes的缩写，就是要获得当前图形的座标轴

# 开始一个循环，把图形文件中各个省的多边形取出来，给它一个颜色，然后把这个多边形放在我们图形的座标轴上
# enumerate() 函数用于将一个可遍历的数据对象(如列表、元组或字符串)组合为一个索引序列，同时列出数据和数据下标

for nshape, seg in enumerate(m.states): # nshape是索引值, seg是元素
    
    color = rgb2hex(colors[statenames[nshape]])  # RGB 与 16进制的 换算，RGB的取值都是0~255，HEX的取值为0123456789ABCDEF这16个字符
    poly = Polygon(seg, facecolor=color, edgecolor=color)  # 把各个省的颜色描在地图上
    
    ax.add_patch(poly)

plt.show()
```
